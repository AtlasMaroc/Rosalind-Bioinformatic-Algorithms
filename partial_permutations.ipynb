```python
class PartialPermutation:
    def __init__(self, n, k):
        self.n = n
        self.k = k 
        self.permutations = perm(n, k)
    
    @staticmethod
    def perm(n, k):
        if k == 0:
            return 1
        return math.perm(n, k)
```
